```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from keras.preprocessing import image
from glob import glob
import matplotlib.image as mpimg
import os
import cv2
```

```python
 #retriving an array of files in directories
TRAINING_DATA = "/content/drive/MyDrive/DatasetSplitted/train"

TEST_DATA = "/content/drive/MyDrive/DatasetSplitted/test"

VALIDATION_DATA = "/content/drive/MyDrive/DatasetSplitted/val"






def plot_images(path, labeled=True, max_images=6):
  amount = 0
  fig = plt.figure(figsize=(10, 6))

  for file in os.listdir(path):
    if file.endswith('.png'):
      if amount == max_images:
        break

      img = mpimg.imread(os.path.join(path, file))
      plt.subplot(231+amount)
      if labeled:
        plt.title(file.split () [1])
      imgplot = plt.imshow(img)

      amount += 1

plot_images(TRAINING_DATA + '/malignant')


plot_images(TRAINING_DATA + '/normal')


plot_images(TRAINING_DATA + '/benign')







```
